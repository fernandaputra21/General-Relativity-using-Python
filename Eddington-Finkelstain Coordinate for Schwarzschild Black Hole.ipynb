```python
#Radial Null Geodesics using numerical integration 
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate
```

The metric function of Schwarzschild black hole is described as
$\newline$
$f(r) = 1 - \frac{2m}{r}$

```python
def f(m,r):
    return 1 - (2*m/r)
```

```python
v = np.arange(0.001,10,0.01)
```

```python
plt.plot(v,f(1,v), 'k')
plt.title('Schwarzschild Black Hole')
plt.xlim(0,2.5)
plt.ylim(-100,100)
```

To see the behavior of space time inside the black hole we need to integrate
$f(r)^{-1}$

```python
def invf(m,r):
    return 1/(1-(2*m/r))
```

```python
plt.plot(v,invf(1,v))
plt.xlim(0,3)
plt.ylim(-100,100)
plt.xlabel('r (distance)')
```

We can see the discontinuity at r = 2.0

```python
def invf(r):
    return 1/(1-(2/r))
```

```python
from scipy import integrate
# Using scipy
upbound = 10
a = 0
e = 0.1
G = []
Hb = []
while e <= upbound:
    if e < 1.99:
        integ = integrate.quad(invf,a,e)[0]
        G.append(e)
        Hb.append(integ)
        e = e + 0.001
        print("CASE 1",e,integ)
    if e > 1.99 and e < 2.001:
        #integ = 100
        #G.append(e)
        #Hb.append(integ)
        e = 2.0011
        print("CASE 2",e,integ)
    if e >= 2.0011:
        integ = integrate.quad(invf,0, 1.99)[0] + integrate.quad(invf, 2.01, e)[0]
        G.append(e)
        Hb.append(integ)
        e = e + 0.001   
        print("CASE 3",e,integ)
```
